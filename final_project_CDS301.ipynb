```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("project_data_32_63.csv")
df.head()
```

Sea ice key
A - ice free/open sea
B - drift ice (loose or very open)
C - loose/open pack ice (pack ice is mobile ice that is not attached to land) 
D - loose/open pack ice with openings 
E - close pack ice with openings 
F - very close pack ice 
G - new ice (brittle or pancakes) 
H - ship is outside of but within sight of pack ice (followed by a direction) 
I - numerous icebergs 
J - several icebergs 
K - very few icebergs 

```python
import warnings
warnings.filterwarnings('ignore')

ice_df = pd.DataFrame(columns=["Year","Month","Date","Letter","Lat","Lon"])

df["split date"] = None
for i in range(len(df)):
    df["split date"][i] = df["date"][i].split("/")
    if type(df["Ice"][i]) == str: 
        if "G" in df["Ice"][i]:
            ice_df.loc[len(ice_df)] = [int(df["split date"][i][2]) + 1900, int(df["split date"][i][0]), int(df["split date"][i][1]), "G", df["Latitude"][i], df["Longitude"][i]]
        if "H" in df["Ice"][i]:
            ice_df.loc[len(ice_df)] = [int(df["split date"][i][2]) + 1900, int(df["split date"][i][0]), int(df["split date"][i][1]), "H", df["Latitude"][i], df["Longitude"][i]]

ice_df
```

```python
h_data = ice_df[(ice_df["Letter"]=="H")]
h_plot = px.scatter_geo(h_data, lat="Lat", lon="Lon", color="Year", projection="azimuthal equal area", title="All Observations of Pack Ice by Year")
h_plot.show()
```

```python
g_nov = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==11)]
g_dec = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==12)]
g_jan = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==1)]
g_feb = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==2)]
g_mar = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==3)]
g_apr = ice_df[(ice_df["Letter"]=="G") & (ice_df["Month"]==4)]

h_nov = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==11)]
h_dec = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==12)]
h_jan = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==1)]
h_feb = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==2)]
h_mar = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==3)]
h_apr = ice_df[(ice_df["Letter"]=="H") & (ice_df["Month"]==4)]
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=3, cols=2, subplot_titles=("November","December","January","February","March"), specs=[[{"type": "scattergeo"}, {"type": "scattergeo"}], [{"type": "scattergeo"}, {"type": "scattergeo"}], [{"type": "scattergeo"}, {"type": "scattergeo"}]])

nov_plot = px.scatter_geo(h_nov, lat="Lat", lon="Lon", color="Year")
dec_plot = px.scatter_geo(h_dec, lat="Lat", lon="Lon", color="Year")
jan_plot = px.scatter_geo(h_jan, lat="Lat", lon="Lon", color="Year")
feb_plot = px.scatter_geo(h_feb, lat="Lat", lon="Lon", color="Year")
mar_plot = px.scatter_geo(h_mar, lat="Lat", lon="Lon", color="Year")
apr_plot = px.scatter_geo(h_apr, lat="Lat", lon="Lon", color="Year")

figure1_traces = []
figure2_traces = []
figure3_traces = []
figure4_traces = []
figure5_traces = []
figure6_traces = []

for trace in range(len(nov_plot["data"])):
    figure1_traces.append(nov_plot["data"][trace])
for trace in range(len(dec_plot["data"])):
    figure2_traces.append(dec_plot["data"][trace])
for trace in range(len(jan_plot["data"])):
    figure3_traces.append(jan_plot["data"][trace])
for trace in range(len(feb_plot["data"])):
    figure4_traces.append(feb_plot["data"][trace])
for trace in range(len(mar_plot["data"])):
    figure5_traces.append(mar_plot["data"][trace])
for trace in range(len(apr_plot["data"])):
    figure6_traces.append(apr_plot["data"][trace])

for traces in figure1_traces:
    fig.append_trace(traces, row=1, col=1)
for traces in figure2_traces:
    fig.append_trace(traces, row=1, col=2)
for traces in figure3_traces:
    fig.append_trace(traces, row=2, col=1)
for traces in figure4_traces:
    fig.append_trace(traces, row=2, col=2)
for traces in figure5_traces:
    fig.append_trace(traces, row=3, col=1)

# April is not plotted because it has no data 

fig.update_layout(autosize=False, width=1000, height=1000, title="Observations of Pack Ice by Month and Year")
fig.show()
```

```python
dec_avg = pd.DataFrame(columns=["Time Span","Avg Lat","Total"])

lats = [[],[],[],[],[],[]]
for i, row in h_dec.iterrows():
    if row["Year"] >= 1935 and row["Year"] <= 1939:
        lats[0].append(row["Lat"])
    elif row["Year"] >= 1940 and row["Year"] <= 1944:
        lats[1].append(row["Lat"])
    elif row["Year"] >= 1945 and row["Year"] <= 1949:
        lats[2].append(row["Lat"])
    elif row["Year"] >= 1950 and row["Year"] <= 1954:
        lats[3].append(row["Lat"])
    elif row["Year"] >= 1955 and row["Year"] <= 1959:
        lats[4].append(row["Lat"])
    elif row["Year"] >= 1960 and row["Year"] <= 1964:
        lats[5].append(row["Lat"])

dec_avg.loc[len(dec_avg)] = ["1935-39",np.average(lats[0]),len(lats[0])]
dec_avg.loc[len(dec_avg)] = ["1940-44",np.average(lats[1]),len(lats[1])]
dec_avg.loc[len(dec_avg)] = ["1945-49",np.average(lats[2]),len(lats[2])]
dec_avg.loc[len(dec_avg)] = ["1950-54",np.average(lats[3]),len(lats[3])]
dec_avg.loc[len(dec_avg)] = ["1955-59",np.average(lats[4]),len(lats[4])]
dec_avg.loc[len(dec_avg)] = ["1960-64",np.average(lats[5]),len(lats[5])]

print(dec_avg)
```
